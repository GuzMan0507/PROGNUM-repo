```python
from math import *
x = float(input("Value of x:"))
y = float(input("Value of y:"))

phi = math.atan2(y, x)

phi_n = phi % (2*pi)
print("PHI radiants:", phi)
angle_deg= degrees(phi)
print("PHI degrees:", angle_deg)


if 0 <= phi_n < pi/2:
    print("1 quadrant") 
elif pi/2 <= phi_n < pi:
    print("2 quadrant") 
elif pi <= phi_n < (3*pi)/4:
    print("3 quadrant")
elif (3*pi)/4 <= phi_n <2*pi:
    print ("4 quadrant")
```
